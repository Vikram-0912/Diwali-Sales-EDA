# ---> Exploratory Data Analysis

## Based on Gender

```python
# Total Orders by Gender

df_gen = df.groupby(['Gender'])[['Orders']].count().reset_index()
df_gen
```

```python
# Total Orders by Gender Barplot

plt.figure(figsize = (3,7))

ax = sns.barplot(x = 'Gender', y = 'Orders', data=df_gen)
ax.bar_label(ax.containers[0], fontsize = 8, padding = 3) #plotting barlabels

plt.xticks(fontsize = 8)
plt.yticks(fontsize = 8)
plt.xlabel('Gender', fontsize = 9)
plt.ylabel('No. of Orders', fontsize = 9)
plt.title('No. of Orders by Gender', fontsize = 11)

plt.show()
```

```python
# Total Sales by Gender

df_gen2 = df.groupby(['Gender'])[['Amount']].sum().reset_index()
df_gen2
```

```python
# Total Sales by Gender Barplot

plt.figure(figsize = (3,6))

ax = sns.barplot(x = 'Gender', y = 'Amount', data=df_gen2)
ax.bar_label(ax.containers[0], fmt='{:,.0f}', fontsize = 7, padding = 3)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}')) #setting the format of y-axis

plt.xticks(fontsize = 8)
plt.yticks(fontsize = 8)
plt.xlabel('Gender', fontsize = 9)
plt.ylabel('Total Sales', fontsize = 9)
plt.title('Sales by Gender', fontsize = 11)

plt.show()
```

```python
# Percentage of Sales by Gender

plt.figure(figsize = (3,7))

ax = sns.histplot(data = df['Gender'], bins = 3, stat = 'percent', color = 'purple')
ax.bar_label(ax.containers[0], fmt='{:,.0f}', fontsize = 8, label_type = 'edge', padding = 5)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.xticks(fontsize = 8)
plt.yticks(fontsize = 8)
plt.xlabel('Gender', fontsize=9)
plt.ylabel('Percentage (%)', fontsize=9)
plt.title('Sales % by Gender', fontsize = 10)

plt.show()
```

From above analysis we can see that most of the buyers are females (70%) and even the purchasing power of females are greater than men
